```python
# Dependencies
import pandas as pd
import os
import csv
import numpy as np
```

```python
# load CSV
csvpath = os.path.join("Resources", "2016-FCC-New-Coders-Survey-Data.csv")
```

```python
# Readfd with pandas

dataclean1_df = pd.read_csv(csvpath,dtype=object)

dataclean1_df.head(10)
```

```python
# Inspect all columns
dataclean1_df.columns
```

```python
# Extract only columns 0, 1, 2, 3, 4, 7, 8, 9, 10,11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
# Use iloc to accomplish this (remember that you need to take the position of the rows and columns into account)

#print (dataclean1_df.dtypes)

new_codevs2_pd = dataclean1_df.iloc[:,[0, 1, 2, 3, 4, 7, 8, 9, 10,11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
new_codevs2_pd.head(5)

```

```python
print (dataclean1_df.dtypes)
```

```python
# Change "0" to "No" and "1" to "Yes" in response columns
# Hint use the df.replace function

#clean_new_coders_pd['AttendedBootcamp'].replace(fixes)

yesno = {"0":'No', "1":'Yes'}
new_codevs2_pd.AttendedBootcamp.replace(yesno, inplace=True)
new_codevs2_pd.head()

#writer = pd.ExcelWriter('new_pdf.xlsx')
#new_codevs2_pd.to_excel(writer,'Sheet1')
#writer.save()
#writer.close()


```

```python
# Calculate total number of respondents in survey
# consider using the len() function
Respondent_count = len(new_codevs2_pd)
Respondent_count

```

```python
# Extract rows corresponding only to people who attended a bootcamp
# Filter using df.loc on the AttendedBootcamp column
Bootcampstay_df = new_codevs2_pd.loc[new_codevs2_pd["AttendedBootcamp"]== "Yes"]
Bootcampstay_df
                                     
```

```python
print (Bootcampstay_df.dtypes)
```

```python
# Calculate average age of attendees
# Pull out the `Age` series and take the average of it

#my_df_2["account_balance"]

Bootcampstay_df['Age'] = pd.to_numeric(Bootcampstay_df['Age'])
Bootcamp_age = Bootcampstay_df["Age"]
Bootcamp_age

```

```python
Avg_age = Bootcampstay_df["Age"].mean()
Avg_age
```

```python
# Calculate how many people attended a bootcamp
# Count the values of the AttendedBootcamp column
AttendedBootcamp_count = Bootcampstay_df["AttendedBootcamp"].count()
AttendedBootcamp_count
```

```python
# Calculate how many attendees hold degrees
# It is easier to do this with `value_counts() count values for the SchoolDegree column

Bootcampstay_df["SchoolDegree"].value_counts()
```

```python
# Count the number of records where the person is a degree holder
# There are several ways to approach this. You can look for people who have degrees
# or for people who don't have degrees depending on the value of the SchoolDegree column

School_Degreescount_df = Bootcampstay_df.loc[(Bootcampstay_df["SchoolDegree"] =="bachelor's degree") |
                                            (Bootcampstay_df["SchoolDegree"] =="master's degree (non-professional)")|
                                            (Bootcampstay_df["SchoolDegree"] =="professional degree (MBA, MD, JD, etc.)")|
                                            (Bootcampstay_df["SchoolDegree"] =="associate's degree")|
                                            (Bootcampstay_df["SchoolDegree"] =="Ph.D."),:]

DegreeCount = School_Degreescount_df["SchoolDegree"].count()

DegreeCount

# 1 way Sri's WBootcampstay_df.loc[Bootcampstay_df["SchoolDegree"] ==""ay
#degree_count = bootcamp_df['SchoolDegree'].loc[(bootcamp_df['SchoolDegree'] == "bachelor's degree"  | (bootcamp_df['SchoolDegree'] == "master\'s degree (non-professional)") | (bootcamp_df['SchoolDegree'] == "professional degree (MBA, MD, JD, etc.)") | (bootcamp_df['SchoolDegree'] == "associate's degree") | (bootcamp_df['SchoolDegree'] == "Ph.D.")].count()

#2 - more efficient than above. 

#degree_count = bootcamp_df['SchoolDegree'].loc[(bootcamp_df['SchoolDegree'].str.contains("degree") == True) & (bootcamp_df['SchoolDegree'].str.contains("no degree") == False) |  (bootcamp_df['SchoolDegree'] == "Ph.D.")].count()                      
                   
#degree_count
```

```python
Gender_combo = {
    'genderqueer':'non-binary',
    'trans':'non-binary'
    ,'agender':'non-binary'
}

BCstayGender_df = Bootcampstay_df.replace({'Gender':Gender_combo})
BCstayGender_df
```

```python
Gender_respond_df=new_codevs2_pd.replace({'Gender':Gender_combo})
```

```python
# Count number of attendees who self-identify as male; female; or are of non-binary gender identification
Gender_respond_df["Gender"].value_counts()
```

```python
# Calculate percentage of respondents who attended a bootcamp

BC_Attend_Pct = (AttendedBootcamp_count/Respondent_count)*100
BC_Attend_Pct
```

```python
# Calculate percentage of respondents belonging to each gender
# Count of each Gender
Male =Gender_respond_df["Gender"].value_counts()['male']
Female = Gender_respond_df["Gender"].value_counts()['female']
NonGender = Gender_respond_df["Gender"].value_counts()['non-binary']

#Calculate percentages

Malepct = (Male/Respondent_count)*100
Femalepct = (Female/Respondent_count) *100
NonGenderpct = (NonGender/Respondent_count)* 100

Malepct
```

```python
# Calculate percentage of attendees with a college degree

Pct_Degreed = DegreeCount/AttendedBootcamp_count*100
Pct_Degreed
```

```python
# Calculate average post-bootcamp salary

Bootcampstay_df['BootcampPostSalary'] = pd.to_numeric(Bootcampstay_df['BootcampPostSalary'])
Camp_salary_df = Bootcampstay_df['BootcampPostSalary']

Pst_btcampsalary = Camp_salary_df.mean()
Pst_btcampsalary
```

```python
# Create a new table consolidating above calculations

BtCamptble = pd.DataFrame({"Total Respondents": [Respondent_count],
                           "Bootcamp Attendees":[AttendedBootcamp_count],
                           "%Attended Bootcamp":[BC_Attend_Pct],
                           "Avg Age":[Avg_age],
                           "Has Degree":[Pct_Degreed],
                           "%Male": [Malepct],
                           "%Female":[Femalepct],
                           "%Non Gender Specific":[NonGenderpct],
                            "Avg Salary": [Pst_btcampsalary]
                            })
BtCamptble = BtCamptble[["Total Respondents","Bootcamp Attendees","%Attended Bootcamp","Avg Age","Has Degree","%Male","%Female","%Non Gender Specific","Avg Salary"]]
BtCamptble=BtCamptble.round(2)
BtCamptble
```
